```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G = nx.cubical_graph()

# Nodes
node_pos = nx.spring_layout(G, seed=3113794652)  # positions for all nodes
red_nodes = [0, 1, 2, 3]
blue_nodes = [4,5,6,7]
options = {"edgecolors": "white", "node_size": 800, "alpha": 0.9}
nx.draw_networkx_nodes(G, node_pos, nodelist=red_nodes, node_color="tab:red", **options)
nx.draw_networkx_nodes(G, node_pos, nodelist=blue_nodes, node_color="tab:blue", **options)

# Edges
red_edges = [(0, 1), (1, 2), (2, 3), (3, 0)]
blue_edges = [(4, 5), (5, 6), (6, 7), (7, 4)]
nx.draw_networkx_edges(G, node_pos, width=1.0, edge_color="white", alpha=0.8)
nx.draw_networkx_edges(G,node_pos,edgelist=red_edges,width=6,alpha=0.4,edge_color="tab:red")
nx.draw_networkx_edges(G,node_pos,edgelist=blue_edges,width=6,alpha=0.4,edge_color="tab:blue")


# Some math labels
labels = {0:r"$a$",1:r"$b$",2:r"$c$",3:r"$d$",4:r"$\alpha$",5:r"$\beta$",6:r"$\gamma$",7:r"$\delta$"}
nx.draw_networkx_labels(G, node_pos, labels, font_size=22, font_color="whitesmoke")

plt.tight_layout()
plt.axis("off")
plt.show()
```
